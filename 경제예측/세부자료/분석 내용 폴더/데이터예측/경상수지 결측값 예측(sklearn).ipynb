```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression  
from sklearn.model_selection import train_test_split 
from sklearn.metrics import mean_squared_error, r2_score 

import matplotlib
import matplotlib.font_manager as fm
import os
font_loc = 'C:/Windows/Fonts/H2PORM.TTF'
font_name = fm.FontProperties(fname=font_loc).get_name()
matplotlib.rc('font',family=font_name)
plt.rcParams['axes.unicode_minus'] = False  # 한글폰트적용시 - 깨짐방지..
```

```python
# 머신러닝 돌릴 데이터 불러오기
df = pd.read_excel('경상수지_머신러닝_전.xlsx')
df = df.drop(columns = ['Unnamed: 0'])
df
```

```python
# 훈련시키고 예측 값 및 결정계수(R^2) 점수

dft = df.iloc[:-1]
dfp = df.iloc[-1:]
xt = dft[dft.columns.difference(['년월', '경상수지'])]
yt = pd.DataFrame(dft['경상수지'])
xp = dfp[dfp.columns.difference(['년월', '경상수지'])]
yp = pd.DataFrame(dfp['경상수지'])

lr = LinearRegression()
lr.fit(xt, yt)

predict = lr.predict(xp)
predict[0], lr.score(xt,yt)
```
