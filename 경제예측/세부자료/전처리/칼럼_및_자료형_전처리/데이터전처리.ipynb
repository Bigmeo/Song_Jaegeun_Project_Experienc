```python
import pandas as pd
```

```python
df = pd.read_excel('./mini_project/new/1.xls', header = 3)
df = df[['월', '금액', '금액.1']]
df.columns = ['월', '수출', '수입']
df = df.sort_index(ascending=False)
df = df.reset_index(drop = True)

for i in range(27):
    df2 = pd.read_excel(f'./mini_project/new/{i+2}.xls', header = 3)
    df2 = df2[['월', '금액', '금액.1']]
    df2.columns = ['월', '수출', '수입']
    df2 = df2.sort_index(ascending=False)
    df2 = df2.reset_index(drop = True)
    df = pd.concat([df, df2])
df
```
